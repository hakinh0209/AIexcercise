```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# %matplotlib inline
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/bt/breast-cancer.csv")
dataset.head(10)
```

```python
dataset = dataset.replace("M" , 1 )
dataset = dataset.replace("B" , 0 )
dataset
```

```python
# make perceptron object (m_perceptron)
m_perceptron = Perceptron(max_iter=10000, verbose=0, random_state=None, fit_intercept=True, eta0=0.001)

# train the perceptron object (m_perceptron)
m_perceptron.fit(dataset[['radius_mean',
 'texture_mean',
 'perimeter_mean',
 'area_mean',
 'smoothness_mean',
 'compactness_mean',
 'concavity_mean',
 'concave points_mean',
 'symmetry_mean',
 'fractal_dimension_mean',
 'radius_se',
 'texture_se',
 'perimeter_se',
 'area_se',
 'smoothness_se',
 'compactness_se',
 'concavity_se',
 'concave points_se',
 'symmetry_se',
 'fractal_dimension_se',
 'radius_worst',
 'texture_worst',
 'perimeter_worst',
 'area_worst',
 'smoothness_worst',
 'compactness_worst',
 'concavity_worst',
 'concave points_worst',
 'symmetry_worst',
 'fractal_dimension_worst']],dataset['diagnosis'])

# print the coefficients matrix and bias
coefficients = m_perceptron.coef_;
bias = m_perceptron.intercept_;
print ("coefficients = " + str(coefficients))
print ("bias = " + str(bias))
```

```python
# run prediction for all the values in the dataset
prediction = m_perceptron.predict(dataset[['radius_mean',
 'texture_mean',
 'perimeter_mean',
 'area_mean',
 'smoothness_mean',
 'compactness_mean',
 'concavity_mean',
 'concave points_mean',
 'symmetry_mean',
 'fractal_dimension_mean',
 'radius_se',
 'texture_se',
 'perimeter_se',
 'area_se',
 'smoothness_se',
 'compactness_se',
 'concavity_se',
 'concave points_se',
 'symmetry_se',
 'fractal_dimension_se',
 'radius_worst',
 'texture_worst',
 'perimeter_worst',
 'area_worst',
 'smoothness_worst',
 'compactness_worst',
 'concavity_worst',
 'concave points_worst',
 'symmetry_worst',
 'fractal_dimension_worst']])
print (prediction)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(prediction, dataset['diagnosis'])
```

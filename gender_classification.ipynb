```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# %matplotlib inline
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/bt/gender_classification_v7.csv")
dataset.head(10)
```

```python
dataset = dataset.replace("Male" , 1 )
dataset = dataset.replace("Female" , 0 )
dataset
```

```python
# make perceptron object (m_perceptron)
m_perceptron = Perceptron(max_iter=10000, verbose=0, random_state=None, fit_intercept=True, eta0=0.001)

# train the perceptron object (m_perceptron)
m_perceptron.fit(dataset[['long_hair', 'forehead_width_cm', 'forehead_height_cm', 'nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long' ]],dataset['gender'])

# print the coefficients matrix and bias
coefficients = m_perceptron.coef_;
bias = m_perceptron.intercept_;
print ("coefficients = " + str(coefficients))
print ("bias = " + str(bias))
```

```python
# run prediction for all the values in the dataset
prediction = m_perceptron.predict(dataset[['long_hair', 'forehead_width_cm', 'forehead_height_cm', 'nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long']])
print (prediction)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(prediction, dataset['gender'])
```
